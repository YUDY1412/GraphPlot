```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import numpy as np

def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x)

x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y) # temperature
```

```python
print(x)
print(y)
print(X)
print(Y)
print(X.size)
print(X.size)
print(Z.size)
print("X.dtype",X.dtype)
print("Y.dtype",Y.dtype)
print("Z.dtype",Z.dtype)
```

```python
plt.contourf(X, Y, Z, 20, cmap='jet')
cbar =plt.colorbar(location='right')
cbar.set_label('Force', fontsize=12)
```
